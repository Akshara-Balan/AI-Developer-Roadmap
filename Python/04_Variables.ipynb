### What is a Variable?

Named container that stores a value.

Think of it like a labeled box.

📦 name → contains "Python"

📦 age → contains 22

Whenever you need the value, you use the variable's name.

### Creating a Variable

In Python, you create a variable suing assignment operator (=).


name = "Alice"

age = 22


Here:

name is the variable name

"Alice" is the value

= assigns the value to the variable

### Printing Variables

```python
name = "Alice"
print (name)
```

```python
age = 22
print (age)
```

### Changing a Variable

A variable's value can be updated.

```python
age = 20
print (age)

age = 22
print (age)
```

The second assignment replaces the old value.

### Variables can store different types of data

##### String

name = "Alice"

##### Integer

age = 22

##### Float

height = 5.4

##### Boolean

is_student = True

### Printing Multiple Variables

You can print several variables in one statement.

```python
name = "Bob"
age = 25
print (name, age)
```

### Variable Naming Rules

Variable can have a short name like x and y or a more descriptive name like age, name.

* Cannot start with a number

* Hyphen is not allowed

* Spaces are not allowed

* Must start with a letter or a underscore

* Can only contain alpha numeric characters and underscores

* Cannot be any of the Python keywords

### Case Sensitivity

Python treats uppercase and lowercase letters as different.

```python
name = "Akshara"
Name = "Rahul"

print(name)
print(Name)
```

### Casting

If you want to specify the data type of a variable, this can be done with casting

```python
x = str(3)    # x will be '3'
y = int(3)    # y will be 3
z = float(3)  # z will be 3.0

print(x,y,z)
```

### Get the Type

You can get the data type of a variable with the type() function

```python
x = 7
y = "John"
print(type(x))
print(type(y))
```

### Multi Words Variable Names

1. Camel Case: myVariableName

2. Pascal Case: MyVariableName

3. Snake Case: my_variable_name

### Assign Multiple Values

```python
# Many Values to Multiple Variables

x, y, z = "Orange", "Banana", "Cherry"
print(x)
print(y)
print(z)
```

```python
# One value to Multiple Variables

x = y = z = "Orange"
print(x)
print(y)
print(z)
```

```python
# Unpack a collection

# If you have a collection of values in a list, tuple etc. Python allows you to extract the values into variables. This is called unpacking.

fruits = ["apple", "banana", "cherry"]
x, y, z = fruits
print(x)
print(y)
print(z)
```

```python
x = "Python "
y = "is "
z = "awesome"
print(x + y + z)
print(x, y, z)
```
