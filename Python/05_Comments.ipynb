### What are comments?

A comment is a note written inside your code that Python ignores.

Comments are used to:

* Explain what the code does.
* Make the code easier to read.
* Leave reminders for yourself.
* Temporarily disable a line of code during testing.

### Single-Line Comments

A single line comment starts with #

```python
# This is a comment

print("Hello, World!")
```

Here, Python ignores: # This is a comment

It only executes print("Hello, World!")


### Commenting your code

Instead of writing without explanation:

```python
name = "Alice"
age = 22

print(name)
print(age)
```

Write:

```python
# Store user's name
name = "Alice"

# Store user's age
age = 22

# Display the values
print(name)
print(age)

# Now anyone reading the code can understand that the variable 'name' stores the user's name and 'age' stores the user's age.
```

### Inline Comments

You can also write comments at the end of a line

```python
age = 22  # User's age
```

### Temporarily Disable Code

During debugging, you can comment out a line so it  doesn't run.

```python
print("Hello")

# print("This line won't run")

print("Python")
```

The commented-out print() statement is ignored.

### Multi-Line Comments

Python doesn't have a special syntax for multi-line comments.

The common approach is to use # on each line.

```python
# This program stores
# the user's details
# and prints them.
```

You'll often see triple quotes (""" ... """) used as documentation strings (docstrings), not regular comments.

```python
"""
This is a docstring.
It can span multiple lines.
"""
```
