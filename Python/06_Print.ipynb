print() function is used to display text or output values.

You can use the function as many times as you want. Each call prints text on a new line by default

Text in Python must be inside quotes. you can use either double quotes or single quotes.

```python
print("This will work!")
print('This will also work!')
```

### Python without a new line

By default, print() function ends witha new line.

If you want to print multiple words on the same line, you can use the end parameter:

```python
print("Hello World!", end=" ")
print("I will print on the same line.")
```

### Print Numbers

```python
# You can use the function to display numbers. We dont put numbers inside quotes

print(7)

# You can also do math

print(3+3)
```

### Mix Text and Numbers

You can combine text and numbers in one output by separating them with a comma:

```python
print("I am", 22, "years old.")
```
