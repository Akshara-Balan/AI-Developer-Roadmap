### What is a String?

A string is a sequence of characters enclosed in single quotes ('), double quotes ("), or triple quotes (''' or """).

```python
name = "Akshara"

city = 'Kochi'

message = "Welcome to Python"
```

All of these are strings.

#### Creating Strings

##### Using Double Quotes

```python
name = "Akshara"

print(name)
```

##### Using Single Quotes

```python
city = 'Kochi'

print(city)
```

##### Using Triple Quotes

Triple quotes are useful for writing text that spans multiple lines.

```python
message = """Welcome
to
Python"""

print(message)
```

#### Accessing Characters (Indexing)

Each character in a string has a position called an index.

Example:

A l i c e

0 1 2 3 4

```python
name = "Akshara"

print(name[0])
```

#### Negative Indexing

Python also counts from the end of the string.

A  k  s  h  a  r  a
-7 -6 -5 -4 -3 -2 -1

```python
name = "Akshara"

print(name[-1])
```

#### String Length

Use the len() function.

```python
name = "Akshara"

print(len(name))
#The length includes every character except the surrounding quotes.
```

### String Concatenation

Concatenation means joining strings.

```python
first_name = "Akshara"
last_name = "Balan"

print(first_name + last_name)
```

```python
# to add space
print(first_name + " " + last_name)
```

### String Repetition

Use *

```python
print("python " * 3)
```

### String Slicing

Slicing lets you extract part of a string.

string[start:end]

The start index is included, but the end index is not.

```python
name = "Happy Learning"

print(name[0:2])
print(name[3:5])
print(name[4:])
```

### Common String Methods

1. upper()- Converts to uppercase

```python
name = "Akshara"
print(name.upper())
```

2. lower()- Converts to lowercase

```python
print(name.lower())
```

3. title()- Capitalizes the first letter of each word.

```python
name = "Bob is smart"
print(name.title())
```

4. capitalize()- Capitalizes only the first letter of first word.

```python
name = "ashna is the heroine"
print(name.capitalize())
```

5. strip()- Removes spaces from the beginning and end.

```python
name = "    People are learning Python    "
print(name.strip())
```

6. replace()- Replaces part of a string

```python
text = "Python is a programming language. Python is easy to learn."
print(text.replace("Python", "Java"))
```

7. find()- Returns the index of the first occurence.

```python
name = "Akshara"
print(name.find('a'))

print(name.find('z'))  # returns -1 if not found
```

8. count()- Counts how many times something appears.

```python
print(name.count('a'))
```

### Checking String Content

startswith() and endswith()

```python
name = "Bob"
print(name.startswith("Bob"))
print(name.startswith("B"))
print(name.startswith("b")) #casesensitive so false
print(name.startswith("o"))
```

```python
print(name.endswith("b")) 
print(name.endswith("B")) #casesensitive so false
print(name.endswith("o"))
print(name.endswith("Bob"))
```

### Escape Characters

Suppose you want to print:

I'm learning Python

Use a backslash (\) to escape the apostrophe:

```python
print('I\'m learning Python')

#or

print("I'm learning Python")
```

### f-Strings

Instead of joining strings with +, use f-strings.

```python
name = "Akshara"
age = 22

print(f"My name is {name} and I am {age} years old.")
```
