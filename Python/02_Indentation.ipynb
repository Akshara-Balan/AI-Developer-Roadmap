Identation means leaving spaces or a tab at the beginning of a line.

Python uses identation to define blocks of code.

##### Why is Indentation important?

In languages like C, Java, or JavaScript, blocks of code are enclosed with curly braces {}.

Example in Java:

if (age >= 18)

{

    System.out.println("Eligible");

}

In python, there are no curly braces. Instead, indentation tells Python which statements belong together.

```python
age = 20

if age >= 18:
    print("Eligible")
```

The print() statement is indented, so Python knows it belongs to the if block.

### Correct Indentation

```python
age = 20

if age >= 18:
    print("You can vote")
```

### Incorrect Indentation

```python
age = 20

if age >= 18:
print("You can vote")
```

Python style guide recommends 4 spaces for eacch indentation level.

Avoid mixing tabs and spaces in the same file.

##### Nested Indentation

You can have blocks inside other blocks.

```python
age = 20

if age >= 18:
    print("Adult")

    if age >= 21:
        print("Can enter the club")
```

Think of it like an outline.

    Chapter

        Section

            Topic

        Another section

Each level is grouped by its indentation.
