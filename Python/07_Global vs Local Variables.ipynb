### What is a Global variable?

Variable that is created outside any function.

Can be accessed from anywhere in the program, including inside functions.

```python
name = "Akshara"   # Global variable

print(name)
```

Here, name is a global variable because it is declared outside any function.

#### Example 1: Accessing a Global Variable inside a Function

```python
name = "Akshara"   # Global variable

def greet():
    print("Hello", name)

greet()
```

The function greet() can access the global variable name, and prints "Hello Akshara".

#### Example 2: Multiple Functions Using the Same Global Variable

```python
college = "ABC College"

def student():
    print(college)

def teacher():
    print(college)

student()
teacher()
```

### Can we change a global variable inside a function?

```python
count = 10

def update():
    count = 20
    print(count)

update()
print(count)
```

Inside the function, Python created a new local variable named count.

The global variable remained unchanged.

### The global Keyword

Ig you want to modify a global variable inside a function, use global keyword.

```python
count = 10

def update():
    global count
    count = 20

update()

print(count)
```

Here:

global count tells Python to use the global variable.

The original value changes from 10 to 20.

### Global vs Local Variable

```python
name = "Akshara"   # Global

def demo():
    city = "Kochi"   # Local
    print(name)
    print(city)

demo()

print(name)

# print(city)  # This would cause an error
```

If you uncomment print(city), you'll get:

```python
name = "Akshara"   # Global

def demo():
    city = "Kochi"   # Local
    print(name)
    print(city)

demo()

print(name)

print(city)  # This would cause an error
```

Because city exists only inside the function.

Imagine your house:

🛏️ Local variable = A notebook kept in your bedroom.

A global variable is declared outside a function.

It can be read inside functions.

To modify a global variable inside a function, use the global keyword.

A local variable exists only inside the function where it is created.

In professional Python code, global variables are used sparingly to keep programs clean and maintainable.
